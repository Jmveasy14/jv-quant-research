```python
import sys
import os

# Add the root directory to the Python path
sys.path.append(os.path.abspath(".."))


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from data_prep import get_returns, get_volatility, to_freq
```

```python
df = pd.read_csv("../sample_data/multi_stock.csv", index_col="Date", parse_dates=True)
price = df  # all 3 tickers included
```

```python
price.plot(title="AAPL Closing Price", figsize=(12, 4))
plt.show()
```

```python
log_rets = get_returns(price, mode="log")
log_rets.plot(title="Log Returns", figsize=(12, 4))
plt.show()
```

```python
vol = get_volatility(log_rets, window=5)
vol.plot(title="22-Day Rolling Volatility", figsize=(12, 4))
plt.show()
```

```python
monthly = to_freq(price, "M")
returns = get_returns(monthly)


sns.heatmap(returns.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap (Monthly Returns)")
plt.show()
#Not showing data because there  is only 1 stock 
```
